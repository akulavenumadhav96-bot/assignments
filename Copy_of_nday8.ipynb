```python
import math

# Get all attributes (including functions) from the math module
math_attributes = dir(math)

# Filter out only the callable attributes (functions)
math_functions = [attr for attr in math_attributes if callable(getattr(math, attr))]

print("Functions available in the 'math' module:")
for func_name in sorted(math_functions):
    print(f"- {func_name}")
```

### Python `sys` Module Functions

The `sys` module provides access to system-specific parameters and functions. It allows you to interact with the Python interpreter itself. Here are some of its functions:

```python
import sys

# Get all attributes (including functions) from the sys module
sys_attributes = dir(sys)

# Filter out only the callable attributes (functions)
sys_functions = [attr for attr in sys_attributes if callable(getattr(sys, attr))]

print("Functions available in the 'sys' module:")
for func_name in sorted(sys_functions):
    print(f"- {func_name}")
```
